```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.DataFrame(df)

# Group data by hue
grouped = df.groupby('class')

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=len(grouped), figsize=(12, 4))

# Iterate over groups and plot KDEs
for i, (name, group) in enumerate(grouped):
    sns.kdeplot(group['class'], ax=axes[i], label=name, shade=True)
    axes[i].set_title(f'KDE for {name}')

plt.tight_layout()
plt.show()
```
